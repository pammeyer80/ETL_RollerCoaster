```python
#Import Libraries
import pandas as pd
from sqlalchemy import create_engine
from config import (user, password, host, port, database)
```

## Extract

We begin by extracting data from each 2 csv files containing roller coaster and amusement park visitation data. The data from each csv is read into 2 dataframes.

The file sources are from kaggle.com:

<b>Roller Coaster Data </b> <br/>
https://www.kaggle.com/lyallstewart/roller-coaster-data

<b>Most Visited Amusement Theme Parks Worldwide 2000 </b><br/>
Timeseries Most Visited Amusement Theme Parks Worldwide 2000-2019.
https://www.kaggle.com/mathurinache/mostvisitedamusementthemeparksworldwide20002019



```python
#Import Roller Coaster file 
coaster_file = "resources/IT_Export.csv"
coaster_df = pd.read_csv(coaster_file)

coaster_df.head()
```

```python
#Import Amusement Park file
park_file = "resources/MostVisitedAmusementThemeParks.csv"
park_df = pd.read_csv(park_file)

park_df.head()
```

## Transform
Next in the transform process we pull desired columns into new dataframes.

For the roller coaster data, we copy columns Coaster, Park, City, State, Type, Year_Opened from the original dataframe into the clean_coaster_df. Then, in order to make sure columns are mapped correctly in the "Load" section, we create a new dataframe lc_clean_coaster_df with columns renamed using lower case letters.

For the worldwide amusement park data, we copy columns Event, Country and 2000-2019 from the original dataframe into clean_park_df. Then proceed to rename the columns with lowercase letters. Event was specifically renamed to "park" to map to the "park" column in the roller coaster data.


```python
#transform and clean up coaster data
#Pull columns Coaster Park City State Type Year_Opened
clean_coaster_df = coaster_df[['Coaster', 'Park', 'City', 'State', 'Type', 'Year_Opened']]
lc_clean_coaster_df = clean_coaster_df.rename(columns={'Coaster':'coaster', 'Park':'park', 'City':'city', 'State':'state', 'Type':'type', 'Year_Opened':'year_opened'})

lc_clean_coaster_df
```

```python
#Transform and clean up the amusement park data
#Pull columns Event, Country and 2000-2019
clean_park_df = park_df[['Event', 'Country', '2000-2019']].copy()

#Rename columns
clean_park_df.rename(columns={"Event": "park", "Country":"country","2000-2019":"visits_2000_2019"}, inplace=True)


#Cleaned up data
clean_park_df
```

## Load
Using the cleaned dataframes lc_clean_coaster_df and clean_park_df, we load each into the database coaster_db. 

lc_clean_coaster_df is loaded into the table "coaster"

clean_park_df is loaded into the table "worldwide"



```python
#Connect to the database
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')
conn = engine.connect()
```

```python
#Verify tables exist in database
engine.table_names()
```

```python
#Load coaster table
lc_clean_coaster_df.to_sql(name='coaster', con=engine, if_exists='append', index=False)

```

```python
#Load worldwide table
clean_park_df.to_sql(name='worldwide', con=engine, if_exists='append', index=False)
```

## Confirm Data Exists

Finally, we confirm that data was loaded into the database tables successfully using select statements against the coaster and worldwide tables. We also identify the records that can be joined together from both tables on the "park" column from each table.

```python
#Confirm data was loaded into coaster table
pd.read_sql_query('select * from coaster', con=engine)
```

```python
#Verify data was loaded into worldwide table
pd.read_sql_query('select * from worldwide', con=engine)
```

```python
#Two Results matched on the park fields in both coaster and worldwide tables
pd.read_sql_query('select * from coaster c inner join worldwide w on c.park = w.park', con=engine)
```
